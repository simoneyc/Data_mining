```python
import math
arr_total = [1/4]*7+[2/4]*5+[3/4]*2
arr_no_gender = [1/3]*9+[2/3]*3
arr_no_depart = [1/3]*5+[2/3]*7
arr_no_salary = [1/3]*7+[2/3]*5
arr_no_mate = [1/3]*10+[2/3]*1

def getE(arr_total):
  agg=0
  for i in range(len(arr_total)):
    a= arr_total[i]
    c= 1-a
    if c == 0:
      c = 1
    agg += -math.log(a,2)*a -math.log(c,2)*c
  return agg

totalE  = getE(arr_total)
genderE =getE(arr_no_gender)
departE =getE(arr_no_depart)
salaryE =getE(arr_no_salary)
mateE  =getE(arr_no_mate)


print('Total:',getE(arr_total))
print('移除性別:',genderE)
print('移除部門:',departE)
print('移除薪資:',salaryE)
print('移除配偶:',mateE)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import manifold,datasets
```

```python
digits = datasets.load_digits(n_class=6)
X, y = digits.data, digits.target
n_saple, n_feature = X.shape
n = 20
img = np.zeros((10*n,10*n))
for i in range(n):
  ix = 10*i+1
  for j in range(n):
    iy = 10*j+1
    img[ix:ix+8,iy:iy+8] = X[i*n+j].reshape((8,8))
  plt.figure(figsize=(8,8))
  plt.imshow(img, cmap=plt.cm.binary)
  plt.xticks([])
  plt.yticks([])
  plt.show()
```

```python
X_tsne = manifold.TSNE(n_components=2, init='random', random_state=5, verbose=1).fit_transform(X)
```

```python
x_min, x_max = X_tsne.min(0),X_tsne.max(0)
X_norm = (X_tsne - x_min) / (x_max - x_min)
plt.figure(figsize=(8,8))
for i in range(X_norm.shape[0]):
  plt.text(X_norm[i,0], X_norm[i,1], str(y[i]), color=plt.cm.Set1(y[i]), fontdict={'weight':'bold','size':9})
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
import pandas as pd
from sklearn import datasets as ds
myiris = ds.load_iris()
```

```python
from sklearn import feature_selection as fs
skb = fs.SelectKBest(fs.f_classif, k=3)
skb.fit(myiris.data, myiris.target)
print(myiris.feature_names)
print(skb.pvalues_)
print(myiris.data[:5])
skb.transform(myiris.data[:5])
```

```python
from sklearn import tree
tc = tree.DecisionTreeClassifier()
tc.fit(myiris.data, myiris.target)
tc.feature_importances_
```

```python
from sklearn import ensemble as es
rfc = es.RandomForestClassifier(n_estimators=1000)
rfc.fit(myiris.data, myiris.target)
rfc.feature_importances_
srm = fs.SelectFromModel(rfc,threshold=0.03,prefit=True)
srm.transform(myiris.data)
```

```python
from sklearn import cluster, datasets
# 讀入鳶尾花資料
iris = datasets.load_iris()
iris_X = iris.data
iris_X
```

```python
# KMeans 演算法
kmeans_fit = cluster.KMeans(n_clusters = 3).fit(iris_X)
```

```python
#印出分群结果
cluster_labels = kmeans_fit.labels_
print("分群結果:")
print(cluster_labels)
print("---")
```

```python
#印出品種看看
iris_y = iris.target
print("真實品種:")
print(iris_y)
```

```python
from sklearn import metrics
# 印出績效
silhouette_avg = metrics.silhouette_score(iris_X, cluster_labels)
print(silhouette_avg)
```

```python
# loop
silhouette_avgs = []
ks = range(2, 11)
for k in ks:
  kmeans_fit = cluster.KMeans(n_clusters = k).fit(iris_X)
  cluster_labels = kmeans_fit.labels_
  silhouette_avg = metrics.silhouette_score(iris_X, cluster_labels)
  silhouette_avgs.append(silhouette_avg)

#作圖並印出k=2到10的績效
import matplotlib.pyplot as plt
plt.bar(ks, silhouette_avgs)
plt.show()
print(silhouette_avgs)
```

```python
# Hierarchical Clustering 演算法
hclust = cluster.AgglomerativeClustering (linkage = 'ward',
                      affinity = 'euclidean',
                      n_clusters = 3)
```

```python
#印出分群結果
hclust.fit(iris_X)
cluster_labels = hclust.labels_
print(cluster_labels)
```

```python
# 印出品種看看
iris_y = iris.target
print("真實品種:")
print(iris_y)
```

```python
from sklearn import metrics
# 印出績效
silhouette_avg = metrics.silhouette_score(iris_X, cluster_labels)
print(silhouette_avg)
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
mergings = linkage(iris_X, method='complete')
dendrogram(mergings,
      labels=iris_y,
      leaf_rotation=90,
      leaf_font_size=6,
      )
plt.show()
```

```python
#安裝套件
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA
```

```python
#建立DBSCAN模型
dbscan = DBSCAN()
dbscan.fit(iris_X)
#使用PCA降維
pca = PCA(n_components=2).fit(iris_X)
pca_2d = pca.transform(iris_X)
```

```python
#依跑出來的分群結果繪圖
for i in range(0, pca_2d.shape[0]):
  if dbscan.labels_[i] == 0:
    c1 = plt.scatter(pca_2d[i, 0], pca_2d[i, 1], c='r', marker='+')
  elif dbscan.labels_[i] == 1:
    c2 = plt.scatter(pca_2d[i, 0], pca_2d[i, 1], c='g', marker='o')
  elif dbscan.labels_[i] == -1:
    c3 = plt.scatter(pca_2d[i, 0], pca_2d[i, 1], c='b', marker='*')
plt.legend([c1, c2, c3], ['Cluster 1', 'Cluster 2', 'Noise'])
plt.title('DBSCAN finds 2 clusters and Noise')
plt.show()
```

```python
# /content/drive/MyDrive/大三下/資料探勘/climate-ds.csv
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/大三下/資料探勘/climate-ds.csv")
df
```

```python
from sklearn.preprocessing import LabelEncoder
# 初始化LabelEncoder
label_encoder = LabelEncoder()

df['Item'] = label_encoder.fit_transform(df['Item'])
df['Area'] = label_encoder.fit_transform(df['Area'])
df
```

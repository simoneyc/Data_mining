```python
import pandas as pd
from sklearn import datasets as ds
```

```python
myiris = ds.load_iris()
```

```python
dfiris = pd.DataFrame(myiris.data,columns=myiris.feature_names);
dfiris_2 = pd.DataFrame(myiris.target,columns=['target']);
```

```python
df_result = pd.concat([dfiris,dfiris_2],axis=1);
df_result.head()
```

```python
from sklearn import preprocessing as pp
print(df_result.mean())
scale = pp.scale(df_result)
```

```python
pp.normalize(df_result)
# scale
```

```python
pp.binarize(scale, threshold=-1)
df_scaled = pd.DataFrame(scale,columns=df_result.columns);
df_scaled
```

```python
pp.binarize(scale,threshold=-1)
df_scaled = pd.DataFrame(scale,columns=df_result.columns);
print(df_scaled.head(3));
dp = df_scaled.applymap(lambda x: 1 if x>-1 else 0)
dp
```

```python
transformer = pp.FunctionTransformer(lambda arr : [[y for y in x] for x in arr])
transformer.transform(scale)[:3]
```
